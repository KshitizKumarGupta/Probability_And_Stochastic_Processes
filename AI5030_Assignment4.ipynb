# **Question:** (UGC-NET, JUNE 2018, Question: 58)
**A simple random sample of size $n$ will be drawn from a class of $125$ students, and the mean mathematics score of the sample will be computed. If the standard error of the sample mean for "*with sampling*" is twice as much as the standard error of the sample mean for "*without replacement sampling*", the value of $n$ is-**


(1.)   32

(2.)   63

(3.)   79

(4.)   94

 ***Ans: (4.) 94***


```python
import numpy as np
from numpy import random
import matplotlib.pyplot as plt
import math
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
# Data related to population

N=125           # given in the question
mean_normal=0   # assumption for the standard normal distribution
```

**Sampling with Replacement :**

---


> $ σ_{\bar{x}}=\frac{σ}{ \sqrt{n}}$

**Sampling without Replacement :**

---


> $ σ_{\bar{x}}=\frac{σ}{ \sqrt{n}}\sqrt{\frac{N-n}{N-1}}   $



where, 

$N$ = Population-Size 

 $n$ = Sample-Size

$σ_{\bar{x}}$ = Standard-Error of the sample population

$σ$ = Standard-Error of the population


Assuming that the population is following a standard normal distribution with mean $μ = 0$ and some standard deviation $σ$, on whose values we form different cases.

# Sample Standard-Error ($σ_{\bar{x}}$) $V/S$ Sample Size ($n$) Plot :

```python
# Cases of considering differnt standard-error of population and analyzing their effect standard-error of sample. 

# CASE I : 
se1=0

# CASE II : 
se2=0.5

# CASE III : 
se3=1

# CASE IV : 
se4=1.5

# CASE V : 
se5=2

# CASE VI : 
se6=100

# CASE VII : 
se7=1000
```

```python
n =np.arange(1,N+1)
```

## CASE I : 

Population's Standard Error ,  $σ=0$

```python
sns.distplot(random.normal(loc=int(mean_normal), scale= se1, size=125),  label='normal')# mean, standard deviation, sample size
plt.show()
```

```python
sample_error_wr=[]
sample_error_wor=[]
for i in n:
  sample_error_wr.append(se1/math.sqrt(i))
  sample_error_wor.append(se1*math.sqrt((N-i)/(N-1))/math.sqrt(i))

plt.plot(n,sample_error_wr,label="With Replacement")
plt.plot(n,sample_error_wor,label="Without Replacement")
plt.legend()
plt.xlabel("Sample Size (n)")
plt.ylabel("Standard Deviation of Sample")
print("Both the graphs have coincided at Standard-Error= 0 for Population.")
plt.show()
```
