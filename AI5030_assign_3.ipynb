# **Question:** (UGC-NET, DEC 2018, Question: 58)
**A sample of size $n(\geq2)$ is drawn from a population of $N\geq3)$ units using $PPSWR$ sampling scheme, where $p_{i}$ is the probability of selecting $i^{th}$ unit in a draw, $0<p_{i}<=1$, $∀i=1,...,N$ and $\sum_{i=1}^{N}p_{i}=1$.**

**Then the inclusion probability $π_{i,j}$ is-**


(1.)   $1-p_{i}^{n}-p_{j}^{n}+(p_{i}+p_{j})^{n}$

(2.)   $1-(p_{i}+p_{j}-p_{i}p_{j})^{n}$

(3.)   $1-(1-p_{i})^{n}-(1-p_{j})^{n}-(p_{i}+p_{j})^{n}$

(4.)    $1-(1-p_{i})^{n}-(1-p_{j})^{n}+(1-p_{i}p_{j})^{n}$



---



 ***Ans: (4.) $1-(1-p_{i})^{n}-(1-p_{j})^{n}+(1-p_{i}p_{j})^{n}$***


### INCLUSION-PROBABILITY ($π_{i,j}$)$ V/S$ PROBABILITY OF SELECTING $i_{th}$ unit IN THE FIRST DRAW ($P_i$) $V/S$ PROBABILITY OF SELECTING jth unit IN THE SECOND DRAW ($P_j$)

> **NOTE :** Sample-size $'n'$ is chosen to be $2$ as per the question but we can change it by changing the value of $n$ given in the following cell.

```python
# SAMPLE-SIZE
n=2 # given in question
```

```python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x, y):
    return (1-(pow((1-x),n)+pow((1-y),n)-pow((1-x*y),n)))

x = np.linspace(0, 1, 30)
y = np.linspace(0, 1, 30)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='binary')
ax.set_xlabel('Pi')
ax.set_ylabel('Pj')
ax.set_zlabel('Inclusion Probability');
```

```python
# TOP VIEW OF THE CONTOUR-PLOT
ax.view_init(60, 35)
fig
```

### Inclusion Probability $V/S$ Sample-size Plot

> **Note:** Population-size $'N'$ remains constant. As per the question, I have used $N=3$ but it can be changed by changing the value of $N$ in the cell below.

```python
# Population-Size
N=3 # given in question
```

```python
incl_prob=[]
sample_size=np.arange(0,N+1)
for n in sample_size:
  incl_prob.append(1-(2*(1-pow((1/N),n))-(1-pow((1-(2/N)),n))))
```

```python
plt.plot(sample_size,incl_prob)
plt.xlabel("Sample Size (n)")
plt.ylabel("Inclusion Probability")
plt.show()
```
