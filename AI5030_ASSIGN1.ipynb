#Question:
**Let $ùëã_1, ùëã_2, ...$ be a sequence of independent normally distributed random variables with
mean $1$ and variance $1$. Let $ùëÅ$ be a Poisson random variable with mean $2$, independent of $ùëã_1, ùëã_2, ...$ . Then the variance of $ùëã_1 + ùëã_2 + ... + ùëã_{ùëÅ +1}$ is -**


(1.)   3

(2.)   4

(3.)   5

(4.)   9

 ***Ans: (1.) 3***



```python
import numpy as np
from numpy import random
import matplotlib.pyplot as plt
import math
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

##SIMULATION OF SEQUENCE OF NORMAL DISTRIBUTIONS

An ideal normal distribution with $mean=2$ and $variance=2$ can be simulated in the below given fashion. 

```python
mean_normal=2
var_normal=2   
```

```python
def normal(mean, var, x):
 temp1=2*math.pi*var
 temp1=pow(temp1,-0.5)
 temp2=pow(x-mean,2)
 temp2=-(temp2)/(2*var)
 return temp1*math.exp(temp2)
```

```python
xcoord=[]
ycoord=[]

for x in np.arange(mean_normal-10, mean_normal+10, 0.01):
    y=normal(mean_normal,var_normal,x)
    xcoord.append(x)
    ycoord.append(y)

plt.plot(xcoord,ycoord)
plt.show()
```

Let's take a real-world simulation of the above case with a sample size of 10000 and plot only finite number of distributions from the infinite sequence of random variables `{Xi's}` for `i= 1 to 2`.  

### SIMULATION OF $X_1$ RANDOM VARIABLE

```python
sns.distplot(random.normal(loc=int(mean_normal), scale=math.sqrt(var_normal), size=10000),  label='normal')# mean, standard deviation, sample size
plt.show()
```

### SIMULATION OF $X_2$ RANDOM VARIABLE

```python
sns.distplot(random.normal(loc=int(mean_normal), scale=math.sqrt(var_normal), size=10000),  label='normal')# mean, standard deviation, sample size
plt.show()
```

## SIMULATION OF POISSON DISTRIBUTION

```python
mean_poisson=2
```

```python
def poisson(mean,x):
  return ((math.exp(-mean)*(mean**x))/math.factorial(x))
```

```python
xcoord=[]
ycoord=[]

for x in np.arange(0, mean_poisson+5, 1):
    y=poisson(mean_poisson,x)
    xcoord.append(x)
    ycoord.append(y)

plt.plot(xcoord,ycoord)
plt.show()
```

```python
sns.distplot(random.poisson(lam=int(mean_poisson), size=10000)) # kde=False
plt.show()
```
